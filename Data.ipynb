```python
totalExposureLog = pd.read_table("totalExposureLog.out", header=None)
```

totalExposureLog, 即历史曝光日志数据，广告n天的曝光历史数据。6.59G，1亿条数据。
（CPC是“Cost Per Click”的英文缩写。在这种模式下广告主仅为用户点击广告的行为付费，而不再为广告的显示次数付费。对 广告主来说,避免了只浏览不点击的风险 ，是网络比较成熟的国家常见的收费方式之一。）
1. 数据请求id：每个请求对应一个用户某一时刻，可能多个广告位
2. 广告请求时间：时间戳，1970年后经过的浮点秒数
3. 广告位id：只区分不同广告位，每个广告位只能曝光特定素材尺寸的广告
4. 用户id
5. 曝光广告id：曝光广告id
6. 曝光广告素材尺寸
7. 曝光广告出价bid：只记录cpc出价，非cpc广告此处记录折算后的cpc价格
8. 曝光广告pctr：预估的pctr，和bid相乘后得到basic_ecpm
9. 曝光广告quality_ecpm: 将广告质量和用户体验等因素折算后的cpc价格
10. 曝光广告totalEcpm：广告排序的分数依据，由basic_ecpm和quality_ecpm相加得到
